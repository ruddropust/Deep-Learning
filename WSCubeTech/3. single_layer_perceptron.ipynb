## What is Perceptron
A perceptron is a neural network unit that does a precise computation to detect features in the input data. Perceptron is mainly used to classify the data into two parts. Therefore, it is also known as Linear Binary Classifier.

The perceptron is a ``single processing unit`` of any neural network. Frank Rosenblatt first proposed in 1958 is a simple neuron which is ``used to classify its input into one or two categories``. Perceptron is a ``linear classifier``, and is ``used in supervised learning``. It helps to organize the given input data.

Perceptron uses the step function that returns +1 if the weighted sum of its input 0 and -1.

The activation function is used to map the input between the required value like (0, 1) or (-1, 1).

### Parts of a Perceptron

A perceptron consists of the following four parts:

1. **Input**: The input values or features of the data.
2. **Weights**: Each input is associated with a weight, which is adjusted during the learning process.
3. **Bias**: An additional parameter that helps the model to fit the data better.
4. **Activation Function**: A function that maps the input to the desired output, typically a step function or a sigmoid function.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.linear_model import Perceptron
```

```python
dataset = pd.read_csv(r"C:\Users\shark\Desktop\Deep Learning\CampusX\Dataset\placement.csv")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="resume_score",hue="placed",data=dataset,legend=2)
```

```python
x = dataset.iloc[:,0:2]
x
```

```python
y = dataset.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=44)
```

```python
pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100, pr.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.values,y.values,clf=pr)
```
