
### Perceptron

The Perceptron is a type of linear classifier, an algorithm for supervised learning of binary classifiers. It is a type of artificial neural network and one of the simplest forms of a neural network model. The Perceptron algorithm learns the weights for the input signals in order to classify the input data into one of two classes.

In this Jupyter Notebook, we have used the Perceptron from the `sklearn.linear_model` module to classify whether a candidate is placed or not based on their CGPA and resume score.

The Perceptron algorithm updates its weights based on the classification errors it makes. For each misclassified example, the weights are adjusted in the direction that would have correctly classified the example. This process is repeated for a specified number of iterations or until the algorithm converges to a solution where all examples are correctly classified. The Perceptron is guaranteed to converge if the data is linearly separable.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv(r"C:\Users\shark\Desktop\Deep Learning\Dataset\placement.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x=df["cgpa"],y=df["resume_score"],hue=df["placed"])
```

```python
sns.pairplot(data=df,hue="placed")
```

```python
x = df.iloc[:,:-1]
y = df["placed"]
```

```python
prc = Perceptron()
prc.fit(x,y)
```

```python
prc.coef_
```

```python
prc.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values, clf=prc,legend=2)
```
