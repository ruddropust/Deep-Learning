```python
import os
import tarfile


# Define the directory where .tgz files are located
tgz_folder = r"J:\train\train\train"  # Change this to your folder path

import os
import subprocess

# Define the directory where .tgz files are located
# tgz_folder = r"path\to\your\tgz\files"  # Change this to your actual folder path
seven_zip_path = r"C:\Program Files\7-Zip\7z.exe"  # Path to 7-Zip executable

# Ensure the folder exists
if not os.path.exists(tgz_folder):
    print("Folder does not exist!")
    exit()

# Loop through all .tgz files in the folder
for file in os.listdir(tgz_folder):
    if file.endswith(".tgz"):
        file_path = os.path.join(tgz_folder, file)
        extract_folder = os.path.join(tgz_folder, os.path.splitext(file)[0])  # Remove .tgz extension

        # Create a folder for extraction if it doesn't exist
        os.makedirs(extract_folder, exist_ok=True)

        # Use 7-Zip to extract
        command = [seven_zip_path, "x", file_path, f"-o{extract_folder}", "-y"]
        subprocess.run(command, check=True)

        print(f"Extracted {file} to {extract_folder}")

print("Extraction completed!")

```
