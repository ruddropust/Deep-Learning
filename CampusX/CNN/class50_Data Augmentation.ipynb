```python
from textwrap import fill

datagen = ImageDataGenerator(
    rotation_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='constant',
)
```

```python
img = image.img_to_array(img)
```

```python
input_batch = img.reshape((1,)+img.shape)
input_batch.shape
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir=r'C:\Users\shark\Desktop\Deep-Learning\CampusX\CNN\aug',save_prefix='cat',save_format='jpeg'):
  i+=1
  if i==10:
    break
```
